##Numpy tutorial

```python
import numpy as np
```

# Creating Array from List

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:" ,arr_1d)
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array: ", arr_2d)
```

## List VS numpy Array

```python
py_list = [1, 2, 3, 4, 5]
print("Python List Multiplication by 2:", py_list * 2)

np_array = np.array([1,2,3])
print("Numpy Array Multiplication by 2:", np_array * 2)

import time 
# Time taken by Python List
start = time.time()
py_list = [i * 2 for i in range(1000000)]
print("\n List operation took: ", time.time() - start)

# Time taken by Numpy Array
start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy operation took: ", time.time() - start)
```

## Creating Array from Scratch

```python
zeros = np.zeros((3,3))
print("Array of Zeros:\n,", zeros)

ones = np.ones((3,3))
print("Array of Ones:\n", ones)

full = np.full((2,3) , 4)
print("Array of Contants\n" , full)

random = np.random.random((3,3))
print("Array of Random\n", random)

sequence = np.arange(0 , 11 , 2)
print("Array of a Sequence\n", sequence)



```

### Vector , Matrix , Tensor

```python
vector = np.array([1,2,3])
print(vector)

Matrix = np.array([[1,2,3],
                   [4,5,6]])
print(Matrix)

tensor = np.array([[[1,2] , [3,4]], [[5,6], [7,8]]])
print(tensor)
```

### Array Properties

```python
arr = np.array([[2,3,4],
                 [3,4,5]])

print("Shape\n", arr.shape)
print("Dimension\n", arr.ndim)
print("size\n", arr.size)
print("Data Type\n", arr.dtype)
```

### Arrays Reshaping


```python
arr = np.arange(12)
print("Original Array: \n",arr)

reshape =   arr.reshape(3,4)
print(reshape)

flattened = reshape.flatten()
print(flattened)

# return views instead of copy
raveled = reshape.ravel()
print(raveled)

#Returns transpose of a MAtrix
transpose = reshape.T
print(transpose)
```
